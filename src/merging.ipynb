```python
import pandas as pd
```

```python
raw1 = {
    "Character":["A","B","C","D","E","F","G"],
    "Number":[1,2,3,4,5,6,7]
}

df1 = pd.DataFrame(raw1)
df1
```

```python
raw2 = {
    "Character":["H","I","J","K","L","M","N"],
    "Number":[8,9,10,11,12,13,14]
}

df2 = pd.DataFrame(raw2)
df2
```

```python
pd.merge(df1,df2,how = "outer",indicator = True)
```

```python
raw3 = {
    "City":["New York","Chicago","Orlando"],
    "Temperature":[21,14,35]
}

df3 = pd.DataFrame(raw3)
df3
```

```python
raw4 = {
    "City":["Chicago","New York","Orlando"],
    "Humidity":[65,68,75]
}

df4 = pd.DataFrame(raw4)
df4
```

```python
pd.merge(df3,df4,on = "City",indicator = True)
```
